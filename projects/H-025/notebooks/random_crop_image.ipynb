```python
from PIL import Image
import random
```

```python
folder = "C:\\Users\\jcruanes\\Desktop\\Images"
image = Image.open(f"{folder}\\image_001.png")
image_label = Image.open(f"{folder}\\label_001.png")
```

```python
for i in range(100):
    random_x = random.randint(0,75-32)
    random_y = random.randint(0,62-32)
    cropped_image = image.crop((random_x, random_y, random_x+32, random_y+32))
    cropped_label = image_label.crop((random_x, random_y, random_x+32, random_y+32))
    cropped_image.save(f"{folder}\\cropped_image_{i:03d}.png")
    cropped_label.save(f"{folder}\\cropped_label_{i:03d}.png")
```
